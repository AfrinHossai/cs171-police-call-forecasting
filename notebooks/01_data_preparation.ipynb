```python
import pandas as pd
import os

file_paths = {
    "2024": os.path.join(data_folder, "police_calls_2024.csv"),
    "2025": os.path.join(data_folder, "police_calls_2025.csv"),
    "2026": os.path.join(data_folder, "police_calls_2026.csv")
}

datasets = {}

for year, path in file_paths.items():
    datasets[year] = pd.read_csv(path, low_memory=False)

    print(f"\n{'=' * 20} {year} {'=' * 20}")
    print("Rows and columns:", datasets[year].shape)
    print("Column names:")
    print(datasets[year].columns.tolist())
```

```python
for year, df in datasets.items():
    print(f"\n{'=' * 25} {year} {'=' * 25}")

    print("\nFirst 3 rows:")
    display(df.head(3))

    print("\nColumn data types:")
    print(df.dtypes)

    print("\nMissing values:")
    missing_values = df.isna().sum()
    print(missing_values[missing_values > 0].sort_values(ascending=False))

    print("\nExact duplicate rows:")
    print(df.duplicated().sum())
```

```python
df_2026 = datasets["2026"]

# Mark every row belonging to an exact-duplicate group
duplicate_mask = df_2026.duplicated(keep=False)
duplicate_rows = df_2026.loc[duplicate_mask].copy()

print("Total 2026 rows:", len(df_2026))
print("Rows belonging to duplicate groups:", duplicate_mask.sum())
print("Extra duplicate copies:", df_2026.duplicated().sum())
print(
    "Unique records represented by duplicate groups:",
    duplicate_rows.drop_duplicates().shape[0]
)

print("\nFirst 20 duplicated rows:")
display(
    duplicate_rows
    .sort_values(["CALL_NUMBER", "EID", "CDTS"])
    .head(20)
)

print("\nLargest duplicate groups by CALL_NUMBER:")
print(
    duplicate_rows
    .groupby("CALL_NUMBER", dropna=False)
    .size()
    .sort_values(ascending=False)
    .head(15)
)
```

```python
cleaned_datasets = {}
duplicate_audit = []

for year, df in datasets.items():
    before_rows = len(df)

    cleaned_df = df.drop_duplicates().reset_index(drop=True)

    after_rows = len(cleaned_df)
    removed_rows = before_rows - after_rows

    cleaned_datasets[year] = cleaned_df

    duplicate_audit.append({
        "year": year,
        "rows_before": before_rows,
        "duplicates_removed": removed_rows,
        "rows_after": after_rows
    })

duplicate_audit_df = pd.DataFrame(duplicate_audit)

display(duplicate_audit_df)

print("\nRemaining exact duplicates:")
for year, df in cleaned_datasets.items():
    print(year, df.duplicated().sum())
```

```python
timestamp_audit = []

for year, df in cleaned_datasets.items():
    df = df.copy()

    # Parse the date and time columns
    parsed_date = pd.to_datetime(
        df["OFFENSE_DATE"],
        errors="coerce",
        format="mixed"
    )

    parsed_time = pd.to_timedelta(
        df["OFFENSE_TIME"].astype(str),
        errors="coerce"
    )

    # Combine date and time into one timestamp
    df["CALL_DATETIME"] = parsed_date.dt.normalize() + parsed_time

    cleaned_datasets[year] = df

    valid_timestamp = df["CALL_DATETIME"].notna()

    timestamp_audit.append({
        "year": year,
        "total_rows": len(df),
        "invalid_dates": parsed_date.isna().sum(),
        "invalid_times": parsed_time.isna().sum(),
        "invalid_timestamps": df["CALL_DATETIME"].isna().sum(),
        "minimum_timestamp": df["CALL_DATETIME"].min(),
        "maximum_timestamp": df["CALL_DATETIME"].max(),
        "timestamps_outside_expected_year": (
            df.loc[valid_timestamp, "CALL_DATETIME"].dt.year != int(year)
        ).sum()
    })

timestamp_audit_df = pd.DataFrame(timestamp_audit)

display(timestamp_audit_df)

print("\nSample timestamps:")
for year, df in cleaned_datasets.items():
    print(f"\n{year}")
    display(
        df[
            ["OFFENSE_DATE", "OFFENSE_TIME", "CALL_DATETIME"]
        ].head(5)
    )
```

```python
combined_parts = []

for year, df in cleaned_datasets.items():
    yearly_df = df.copy()
    yearly_df["SOURCE_YEAR"] = int(year)
    combined_parts.append(yearly_df)

combined_df = pd.concat(
    combined_parts,
    ignore_index=True
)

combined_df = (
    combined_df
    .sort_values("CALL_DATETIME")
    .reset_index(drop=True)
)

print("Combined shape:", combined_df.shape)
print("Expected rows:", sum(len(df) for df in cleaned_datasets.values()))
print("Actual rows:", len(combined_df))

print("\nRows by source year:")
print(
    combined_df["SOURCE_YEAR"]
    .value_counts()
    .sort_index()
)

print("\nTimestamp range:")
print("Minimum:", combined_df["CALL_DATETIME"].min())
print("Maximum:", combined_df["CALL_DATETIME"].max())

print("\nValidation checks:")
print(
    "Chronologically sorted:",
    combined_df["CALL_DATETIME"].is_monotonic_increasing
)
print(
    "Missing timestamps:",
    combined_df["CALL_DATETIME"].isna().sum()
)
print(
    "Year mismatches:",
    (
        combined_df["CALL_DATETIME"].dt.year
        != combined_df["SOURCE_YEAR"]
    ).sum()
)
print(
    "Exact duplicates after combination:",
    combined_df.duplicated().sum()
)

print("\nFirst 3 combined rows:")
display(combined_df.head(3))

print("\nLast 3 combined rows:")
display(combined_df.tail(3))
```

```python
# Assign every incident to its corresponding hourly interval
combined_df["HOUR"] = combined_df["CALL_DATETIME"].dt.floor("h")

# Count incidents in each observed hour
observed_hourly_counts = (
    combined_df
    .groupby("HOUR")
    .size()
    .rename("TOTAL_CALLS")
)

# Create a complete hourly timeline
complete_hour_index = pd.date_range(
    start=combined_df["HOUR"].min(),
    end=combined_df["HOUR"].max(),
    freq="h"
)

# Insert any missing hours with zero calls
hourly_df = (
    observed_hourly_counts
    .reindex(complete_hour_index, fill_value=0)
    .rename_axis("HOUR")
    .reset_index()
)

print("Incident rows in combined data:", len(combined_df))
print("Calls represented in hourly data:", hourly_df["TOTAL_CALLS"].sum())
print("Observed hourly intervals:", len(observed_hourly_counts))
print("Complete hourly intervals:", len(hourly_df))
print(
    "Hours inserted with zero calls:",
    (hourly_df["TOTAL_CALLS"] == 0).sum()
)

print("\nHourly timestamp range:")
print("Minimum:", hourly_df["HOUR"].min())
print("Maximum:", hourly_df["HOUR"].max())

print("\nHourly call-count summary:")
display(hourly_df["TOTAL_CALLS"].describe().to_frame())

print("\nFirst 10 hourly rows:")
display(hourly_df.head(10))

print("\nLast 10 hourly rows:")
display(hourly_df.tail(10))
```

```python
zero_hours = hourly_df.loc[
    hourly_df["TOTAL_CALLS"] == 0,
    "HOUR"
].tolist()

print("Number of zero-call hours:", len(zero_hours))
print("\nZero-call timestamps:")

for zero_hour in zero_hours:
    print(zero_hour)

print("\nSurrounding hourly counts:")

for zero_hour in zero_hours:
    print(f"\n{'=' * 60}")
    print("Zero-call hour:", zero_hour)

    surrounding_hours = hourly_df.loc[
        hourly_df["HOUR"].between(
            zero_hour - pd.Timedelta(hours=3),
            zero_hour + pd.Timedelta(hours=3)
        )
    ]

    display(surrounding_hours)

    incident_count = (
        combined_df["HOUR"] == zero_hour
    ).sum()

    print(
        "Incident-level records during this hour:",
        incident_count
    )
```

```python
unusual_zero_hour = pd.Timestamp("2026-03-17 04:00:00")

print("Hourly counts around the unusual zero hour:")
display(
    hourly_df.loc[
        hourly_df["HOUR"].between(
            unusual_zero_hour - pd.Timedelta(hours=6),
            unusual_zero_hour + pd.Timedelta(hours=6)
        )
    ]
)

print("\nIncident timestamps immediately before the gap:")
display(
    combined_df.loc[
        combined_df["CALL_DATETIME"] < unusual_zero_hour,
        ["CALL_DATETIME", "CALL_NUMBER", "CALL_TYPE", "PRIORITY"]
    ].tail(10)
)

print("\nIncident timestamps immediately after the gap:")
display(
    combined_df.loc[
        combined_df["CALL_DATETIME"] >= unusual_zero_hour
        + pd.Timedelta(hours=1),
        ["CALL_DATETIME", "CALL_NUMBER", "CALL_TYPE", "PRIORITY"]
    ].head(10)
)

print("\nCounts for 4:00 a.m. on nearby dates:")
nearby_4am = hourly_df.loc[
    (hourly_df["HOUR"].dt.hour == 4)
    & hourly_df["HOUR"].between(
        unusual_zero_hour - pd.Timedelta(days=7),
        unusual_zero_hour + pd.Timedelta(days=7)
    )
]

display(nearby_4am)
```

```python
# Preserve the original hourly_df
hourly_corrected_df = hourly_df.copy()

dst_nonexistent_hours = pd.to_datetime([
    "2024-03-10 02:00:00",
    "2025-03-09 02:00:00",
    "2026-03-08 02:00:00"
])

suspected_missing_hour = pd.Timestamp("2026-03-17 04:00:00")

# Remove hours that did not exist because of daylight-saving time
hourly_corrected_df = hourly_corrected_df.loc[
    ~hourly_corrected_df["HOUR"].isin(dst_nonexistent_hours)
].copy()

# Mark the suspected data gap as missing, not as zero calls
hourly_corrected_df["MISSING_DATA_FLAG"] = 0

missing_mask = (
    hourly_corrected_df["HOUR"] == suspected_missing_hour
)

hourly_corrected_df.loc[
    missing_mask,
    "TOTAL_CALLS"
] = pd.NA

hourly_corrected_df.loc[
    missing_mask,
    "MISSING_DATA_FLAG"
] = 1

hourly_corrected_df["TOTAL_CALLS"] = (
    hourly_corrected_df["TOTAL_CALLS"]
    .astype("Int64")
)

hourly_corrected_df = (
    hourly_corrected_df
    .reset_index(drop=True)
)

print("Original hourly rows:", len(hourly_df))
print("Corrected hourly rows:", len(hourly_corrected_df))
print(
    "Remaining zero-call hours:",
    (hourly_corrected_df["TOTAL_CALLS"] == 0).sum()
)
print(
    "Missing hourly values:",
    hourly_corrected_df["TOTAL_CALLS"].isna().sum()
)
print(
    "Calls preserved:",
    hourly_corrected_df["TOTAL_CALLS"].sum()
)

print("\nSuspected missing-data row:")
display(
    hourly_corrected_df.loc[
        hourly_corrected_df["MISSING_DATA_FLAG"] == 1
    ]
)
```

```python
modeling_df = hourly_corrected_df.copy()

# Target: number of calls in the following hourly row
modeling_df["NEXT_HOUR_CALLS"] = (
    modeling_df["TOTAL_CALLS"].shift(-1)
)

# Confirm that the next row represents the expected next clock hour
modeling_df["NEXT_TIMESTAMP"] = (
    modeling_df["HOUR"].shift(-1)
)

modeling_df["HOURS_TO_NEXT"] = (
    modeling_df["NEXT_TIMESTAMP"] - modeling_df["HOUR"]
).dt.total_seconds() / 3600

print("Total rows:", len(modeling_df))
print(
    "Missing current-hour call counts:",
    modeling_df["TOTAL_CALLS"].isna().sum()
)
print(
    "Missing target values:",
    modeling_df["NEXT_HOUR_CALLS"].isna().sum()
)

print("\nTime differences between consecutive rows:")
print(
    modeling_df["HOURS_TO_NEXT"]
    .value_counts(dropna=False)
    .sort_index()
)

print("\nRows with missing current or target values:")
display(
    modeling_df.loc[
        modeling_df["TOTAL_CALLS"].isna()
        | modeling_df["NEXT_HOUR_CALLS"].isna(),
        [
            "HOUR",
            "TOTAL_CALLS",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS",
            "MISSING_DATA_FLAG",
            "HOURS_TO_NEXT"
        ]
    ]
)

print("\nFirst five target examples:")
display(
    modeling_df[
        ["HOUR", "TOTAL_CALLS", "NEXT_TIMESTAMP", "NEXT_HOUR_CALLS"]
    ].head()
)
```

```python
invalid_modeling_rows = modeling_df.loc[
    modeling_df["TOTAL_CALLS"].isna()
    | modeling_df["NEXT_HOUR_CALLS"].isna()
    | (modeling_df["HOURS_TO_NEXT"] != 1)
].copy()

print("Rows excluded from modeling:", len(invalid_modeling_rows))

display(
    invalid_modeling_rows[
        [
            "HOUR",
            "TOTAL_CALLS",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS",
            "MISSING_DATA_FLAG",
            "HOURS_TO_NEXT"
        ]
    ]
)

prepared_df = modeling_df.loc[
    modeling_df["TOTAL_CALLS"].notna()
    & modeling_df["NEXT_HOUR_CALLS"].notna()
    & (modeling_df["HOURS_TO_NEXT"] == 1)
].copy()

prepared_df["TOTAL_CALLS"] = prepared_df["TOTAL_CALLS"].astype(int)
prepared_df["NEXT_HOUR_CALLS"] = prepared_df["NEXT_HOUR_CALLS"].astype(int)

prepared_df = prepared_df.reset_index(drop=True)

print("\nPrepared modeling rows:", len(prepared_df))
print("Missing current counts:", prepared_df["TOTAL_CALLS"].isna().sum())
print("Missing targets:", prepared_df["NEXT_HOUR_CALLS"].isna().sum())
print(
    "Non-one-hour prediction gaps:",
    (prepared_df["HOURS_TO_NEXT"] != 1).sum()
)

print("\nPrepared dataset range:")
print("First prediction hour:", prepared_df["HOUR"].min())
print("Last prediction hour:", prepared_df["HOUR"].max())

display(prepared_df.head())
```

```python
import numpy as np

feature_df = prepared_df.copy()

target_time = feature_df["NEXT_TIMESTAMP"]

# Basic calendar features for the hour being predicted
feature_df["TARGET_HOUR"] = target_time.dt.hour
feature_df["TARGET_DAY_OF_WEEK"] = target_time.dt.dayofweek  # Monday=0
feature_df["TARGET_MONTH"] = target_time.dt.month
feature_df["TARGET_IS_WEEKEND"] = (
    feature_df["TARGET_DAY_OF_WEEK"] >= 5
).astype(int)

# Cyclical encodings
feature_df["HOUR_SIN"] = np.sin(
    2 * np.pi * feature_df["TARGET_HOUR"] / 24
)
feature_df["HOUR_COS"] = np.cos(
    2 * np.pi * feature_df["TARGET_HOUR"] / 24
)

feature_df["DAY_OF_WEEK_SIN"] = np.sin(
    2 * np.pi * feature_df["TARGET_DAY_OF_WEEK"] / 7
)
feature_df["DAY_OF_WEEK_COS"] = np.cos(
    2 * np.pi * feature_df["TARGET_DAY_OF_WEEK"] / 7
)

feature_df["MONTH_SIN"] = np.sin(
    2 * np.pi * (feature_df["TARGET_MONTH"] - 1) / 12
)
feature_df["MONTH_COS"] = np.cos(
    2 * np.pi * (feature_df["TARGET_MONTH"] - 1) / 12
)

calendar_columns = [
    "TARGET_HOUR",
    "TARGET_DAY_OF_WEEK",
    "TARGET_MONTH",
    "TARGET_IS_WEEKEND",
    "HOUR_SIN",
    "HOUR_COS",
    "DAY_OF_WEEK_SIN",
    "DAY_OF_WEEK_COS",
    "MONTH_SIN",
    "MONTH_COS"
]

print("Feature dataset shape:", feature_df.shape)
print(
    "Missing calendar-feature values:",
    feature_df[calendar_columns].isna().sum().sum()
)

print("\nCalendar feature ranges:")
display(feature_df[calendar_columns].agg(["min", "max"]))

print("\nFirst 10 calendar-feature examples:")
display(
    feature_df[
        [
            "HOUR",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS"
        ] + calendar_columns
    ].head(10)
)
```

```python
# Lookup table containing call counts for each actual clock hour
call_count_lookup = (
    hourly_corrected_df
    .set_index("HOUR")["TOTAL_CALLS"]
)

# Lag times are measured relative to the predicted hour
lag_hours = [1, 2, 3, 24, 48, 168]

lag_columns = []

for lag in lag_hours:
    column_name = f"LAG_{lag}H"

    source_timestamp = (
        feature_df["NEXT_TIMESTAMP"]
        - pd.Timedelta(hours=lag)
    )

    feature_df[column_name] = source_timestamp.map(
        call_count_lookup
    )

    lag_columns.append(column_name)

print("Lag columns created:")
print(lag_columns)

print("\nMissing values by lag feature:")
print(feature_df[lag_columns].isna().sum())

print(
    "\nLAG_1H differs from current TOTAL_CALLS:",
    (
        feature_df["LAG_1H"]
        != feature_df["TOTAL_CALLS"]
    ).fillna(False).sum()
)

print("\nFirst 10 lag-feature examples:")
display(
    feature_df[
        [
            "HOUR",
            "TOTAL_CALLS",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS"
        ] + lag_columns
    ].head(10)
)

print("\nRows with any missing lag values:")
display(
    feature_df.loc[
        feature_df[lag_columns].isna().any(axis=1),
        [
            "HOUR",
            "NEXT_TIMESTAMP"
        ] + lag_columns
    ].head(20)
)
```

```python
# Build exact historical hourly values before each predicted timestamp
maximum_history = 168

history_lag_matrix = pd.DataFrame({
    lag: (
        feature_df["NEXT_TIMESTAMP"]
        - pd.Timedelta(hours=lag)
    ).map(call_count_lookup).astype("float64")
    for lag in range(1, maximum_history + 1)
})

rolling_windows = [3, 6, 24, 168]
rolling_columns = []

for window in rolling_windows:
    column_name = f"ROLLING_MEAN_{window}H"

    # Require every exact hour in the window to be available
    feature_df[column_name] = (
        history_lag_matrix
        .loc[:, 1:window]
        .mean(axis=1, skipna=False)
    )

    rolling_columns.append(column_name)

print("Rolling-average columns created:")
print(rolling_columns)

print("\nMissing values by rolling feature:")
print(feature_df[rolling_columns].isna().sum())

# Confirm that the 3-hour rolling mean uses only the three prior hours
complete_3h_mask = feature_df[
    ["LAG_1H", "LAG_2H", "LAG_3H"]
].notna().all(axis=1)

manual_3h_mean = (
    feature_df.loc[
        complete_3h_mask,
        ["LAG_1H", "LAG_2H", "LAG_3H"]
    ]
    .astype(float)
    .mean(axis=1)
)

maximum_difference = (
    feature_df.loc[
        complete_3h_mask,
        "ROLLING_MEAN_3H"
    ]
    - manual_3h_mean
).abs().max()

print(
    "\nMaximum difference in 3-hour validation:",
    maximum_difference
)

print("\nFirst 10 rolling-feature examples:")
display(
    feature_df[
        [
            "HOUR",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS"
        ]
        + lag_columns
        + rolling_columns
    ].head(10)
)

print("\nFirst rows with complete rolling features:")
display(
    feature_df.loc[
        feature_df[rolling_columns].notna().all(axis=1),
        [
            "HOUR",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS"
        ]
        + rolling_columns
    ].head(10)
)
```

```python
print("Priority column data type:", combined_df["PRIORITY"].dtype)
print("Missing priority values:", combined_df["PRIORITY"].isna().sum())

print("\nUnique priority values:")
print(
    sorted(
        combined_df["PRIORITY"]
        .dropna()
        .unique()
        .tolist()
    )
)

print("\nOverall priority distribution:")
priority_distribution = (
    combined_df["PRIORITY"]
    .value_counts(dropna=False)
    .sort_index()
    .rename_axis("PRIORITY")
    .reset_index(name="CALL_COUNT")
)

priority_distribution["PERCENT"] = (
    priority_distribution["CALL_COUNT"]
    / len(combined_df)
    * 100
)

display(priority_distribution)

print("\nPriority distribution by source year:")
priority_by_year = pd.crosstab(
    combined_df["SOURCE_YEAR"],
    combined_df["PRIORITY"],
    dropna=False
)

display(priority_by_year)

print("\nSample records for each priority:")
priority_samples = (
    combined_df
    .sort_values("CALL_DATETIME")
    .groupby("PRIORITY", dropna=False)
    .head(3)
)

display(
    priority_samples[
        [
            "CALL_DATETIME",
            "CALL_NUMBER",
            "PRIORITY",
            "CALL_TYPE"
        ]
    ]
)
```

```python
# Group rare priority 8 into an OTHER category
priority_source = combined_df[
    ["HOUR", "PRIORITY"]
].copy()

priority_source["PRIORITY_GROUP"] = (
    priority_source["PRIORITY"]
    .where(
        priority_source["PRIORITY"].isin([1, 2, 3, 4, 5, 6]),
        "OTHER"
    )
    .astype(str)
)

# Count calls of each priority during each hour
hourly_priority_counts = (
    priority_source
    .groupby(["HOUR", "PRIORITY_GROUP"])
    .size()
    .unstack(fill_value=0)
)

required_priority_groups = ["1", "2", "3", "4", "5", "6", "OTHER"]

hourly_priority_counts = hourly_priority_counts.reindex(
    columns=required_priority_groups,
    fill_value=0
)

hourly_priority_counts.columns = [
    f"CURRENT_PRIORITY_{priority}_COUNT"
    for priority in required_priority_groups
]

priority_columns = hourly_priority_counts.columns.tolist()

# Join counts using the current prediction-origin hour
feature_df = feature_df.merge(
    hourly_priority_counts.reset_index(),
    on="HOUR",
    how="left",
    validate="many_to_one"
)

feature_df[priority_columns] = (
    feature_df[priority_columns]
    .fillna(0)
    .astype(int)
)

# Validate that priority counts equal the current hour's total calls
feature_df["PRIORITY_COUNT_SUM"] = (
    feature_df[priority_columns].sum(axis=1)
)

print("Priority features created:")
print(priority_columns)

print(
    "\nMissing priority-feature values:",
    feature_df[priority_columns].isna().sum().sum()
)

print(
    "Rows where priority counts do not equal TOTAL_CALLS:",
    (
        feature_df["PRIORITY_COUNT_SUM"]
        != feature_df["TOTAL_CALLS"]
    ).sum()
)

print("\nPriority-feature summary:")
display(feature_df[priority_columns].describe())

print("\nFirst 10 priority-feature examples:")
display(
    feature_df[
        [
            "HOUR",
            "TOTAL_CALLS",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS"
        ]
        + priority_columns
        + ["PRIORITY_COUNT_SUM"]
    ].head(10)
)
```

```python
split_df = feature_df.copy()

train_end = pd.Timestamp("2025-09-30 23:00:00")
validation_end = pd.Timestamp("2026-02-28 23:00:00")

split_df["SPLIT"] = np.select(
    [
        split_df["NEXT_TIMESTAMP"] <= train_end,
        (
            (split_df["NEXT_TIMESTAMP"] > train_end)
            & (split_df["NEXT_TIMESTAMP"] <= validation_end)
        ),
        split_df["NEXT_TIMESTAMP"] > validation_end
    ],
    [
        "train",
        "validation",
        "test"
    ],
    default="unassigned"
)

train_df = split_df.loc[
    split_df["SPLIT"] == "train"
].copy()

validation_df = split_df.loc[
    split_df["SPLIT"] == "validation"
].copy()

test_df = split_df.loc[
    split_df["SPLIT"] == "test"
].copy()

split_audit = (
    split_df
    .groupby("SPLIT")
    .agg(
        rows=("NEXT_HOUR_CALLS", "size"),
        prediction_start=("NEXT_TIMESTAMP", "min"),
        prediction_end=("NEXT_TIMESTAMP", "max")
    )
    .reset_index()
)

split_audit["percentage"] = (
    split_audit["rows"]
    / len(split_df)
    * 100
)

display(split_audit)

print("Total rows:", len(split_df))
print("Train rows:", len(train_df))
print("Validation rows:", len(validation_df))
print("Test rows:", len(test_df))

print(
    "\nUnassigned rows:",
    (split_df["SPLIT"] == "unassigned").sum()
)

print("\nChronological validation:")
print(
    "Train ends before validation begins:",
    train_df["NEXT_TIMESTAMP"].max()
    < validation_df["NEXT_TIMESTAMP"].min()
)

print(
    "Validation ends before test begins:",
    validation_df["NEXT_TIMESTAMP"].max()
    < test_df["NEXT_TIMESTAMP"].min()
)

print("\nBoundary rows:")
display(
    split_df.loc[
        split_df["NEXT_TIMESTAMP"].between(
            pd.Timestamp("2025-09-30 21:00:00"),
            pd.Timestamp("2025-10-01 02:00:00")
        ),
        [
            "HOUR",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS",
            "SPLIT"
        ]
    ]
)

display(
    split_df.loc[
        split_df["NEXT_TIMESTAMP"].between(
            pd.Timestamp("2026-02-28 21:00:00"),
            pd.Timestamp("2026-03-01 02:00:00")
        ),
        [
            "HOUR",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS",
            "SPLIT"
        ]
    ]
)
```

```python
# Keep only incident records belonging to training prediction-origin hours
training_hours = pd.Index(train_df["HOUR"].unique())

training_incidents = combined_df.loc[
    combined_df["HOUR"].isin(training_hours)
].copy()

# Create standardized call-type labels for inspection
raw_call_types = training_incidents["CALL_TYPE"].astype("string")

training_incidents["CALL_TYPE_CLEAN"] = (
    raw_call_types
    .str.strip()
    .str.upper()
    .replace("", pd.NA)
)

print("Training hourly rows:", len(train_df))
print("Training incident records:", len(training_incidents))
print(
    "Missing call types:",
    training_incidents["CALL_TYPE_CLEAN"].isna().sum()
)
print(
    "Unique cleaned call types:",
    training_incidents["CALL_TYPE_CLEAN"].nunique()
)

spacing_issues = (
    raw_call_types.fillna("")
    != raw_call_types.fillna("").str.strip()
).sum()

print("Labels with leading/trailing spaces:", spacing_issues)

# Frequency distribution based only on training data
call_type_distribution = (
    training_incidents["CALL_TYPE_CLEAN"]
    .value_counts(dropna=False)
    .rename_axis("CALL_TYPE_CLEAN")
    .reset_index(name="CALL_COUNT")
)

call_type_distribution["PERCENT"] = (
    call_type_distribution["CALL_COUNT"]
    / len(training_incidents)
    * 100
)

call_type_distribution["CUMULATIVE_PERCENT"] = (
    call_type_distribution["PERCENT"].cumsum()
)

print("\nTop 30 training call types:")
display(call_type_distribution.head(30))

# Compare the most common categories across training years
top_20_call_types = (
    call_type_distribution
    .dropna(subset=["CALL_TYPE_CLEAN"])
    .head(20)["CALL_TYPE_CLEAN"]
    .tolist()
)

call_type_by_year = pd.crosstab(
    training_incidents["SOURCE_YEAR"],
    training_incidents["CALL_TYPE_CLEAN"]
)

call_type_by_year = call_type_by_year.reindex(
    columns=top_20_call_types,
    fill_value=0
)

print("\nTop 20 call types by training year:")
display(call_type_by_year)
```

```python
import re

# The top 20 categories were selected using training data only
selected_call_types = top_20_call_types.copy()

print("Selected training call types:")
for index, call_type in enumerate(selected_call_types, start=1):
    print(f"{index:2d}. {call_type}")

selected_coverage = (
    call_type_distribution
    .head(20)["PERCENT"]
    .sum()
)

print(
    "\nTraining-data coverage of selected call types:",
    round(selected_coverage, 2),
    "%"
)

# Clean call-type labels across the complete incident dataset
call_type_source = combined_df[
    ["HOUR", "CALL_TYPE"]
].copy()

call_type_source["CALL_TYPE_CLEAN"] = (
    call_type_source["CALL_TYPE"]
    .astype("string")
    .str.strip()
    .str.upper()
    .replace("", pd.NA)
)

# Preserve the selected training categories and group everything else
call_type_source["CALL_TYPE_GROUP"] = (
    call_type_source["CALL_TYPE_CLEAN"]
    .where(
        call_type_source["CALL_TYPE_CLEAN"].isin(
            selected_call_types
        ),
        "OTHER"
    )
    .fillna("OTHER")
)

# Count each call-type group during every current hour
hourly_call_type_counts = (
    call_type_source
    .groupby(["HOUR", "CALL_TYPE_GROUP"])
    .size()
    .unstack(fill_value=0)
)

required_call_type_groups = (
    selected_call_types + ["OTHER"]
)

hourly_call_type_counts = (
    hourly_call_type_counts
    .reindex(
        columns=required_call_type_groups,
        fill_value=0
    )
)

# Create safe machine-learning column names
call_type_name_map = {}

for call_type in required_call_type_groups:
    safe_name = re.sub(
        r"[^A-Z0-9]+",
        "_",
        call_type
    ).strip("_")

    call_type_name_map[call_type] = (
        f"CURRENT_CALLTYPE_{safe_name}_COUNT"
    )

hourly_call_type_counts = (
    hourly_call_type_counts
    .rename(columns=call_type_name_map)
)

call_type_columns = (
    hourly_call_type_counts.columns.tolist()
)

# Join using the current hour, not the predicted hour
feature_df = feature_df.merge(
    hourly_call_type_counts.reset_index(),
    on="HOUR",
    how="left",
    validate="many_to_one"
)

feature_df[call_type_columns] = (
    feature_df[call_type_columns]
    .fillna(0)
    .astype(int)
)

# Validate that every incident belongs to exactly one category
feature_df["CALL_TYPE_COUNT_SUM"] = (
    feature_df[call_type_columns]
    .sum(axis=1)
)

print("\nNumber of call-type features:", len(call_type_columns))

print(
    "Missing call-type feature values:",
    feature_df[call_type_columns]
    .isna()
    .sum()
    .sum()
)

print(
    "Rows where call-type counts do not equal TOTAL_CALLS:",
    (
        feature_df["CALL_TYPE_COUNT_SUM"]
        != feature_df["TOTAL_CALLS"]
    ).sum()
)

print("\nFirst 10 call-type feature examples:")
display(
    feature_df[
        [
            "HOUR",
            "TOTAL_CALLS",
            "NEXT_TIMESTAMP",
            "NEXT_HOUR_CALLS"
        ]
        + call_type_columns
        + ["CALL_TYPE_COUNT_SUM"]
    ].head(10)
)
```

```python
# Require all lag and rolling-history features to be available
history_feature_columns = lag_columns + rolling_columns

complete_history_mask = (
    feature_df[history_feature_columns]
    .notna()
    .all(axis=1)
)

final_feature_df = (
    feature_df.loc[complete_history_mask]
    .copy()
    .reset_index(drop=True)
)

# Recreate the chronological split because feature_df was updated
final_feature_df["SPLIT"] = np.select(
    [
        final_feature_df["NEXT_TIMESTAMP"] <= train_end,
        (
            (final_feature_df["NEXT_TIMESTAMP"] > train_end)
            & (final_feature_df["NEXT_TIMESTAMP"] <= validation_end)
        ),
        final_feature_df["NEXT_TIMESTAMP"] > validation_end
    ],
    [
        "train",
        "validation",
        "test"
    ],
    default="unassigned"
)

final_train_df = final_feature_df.loc[
    final_feature_df["SPLIT"] == "train"
].copy()

final_validation_df = final_feature_df.loc[
    final_feature_df["SPLIT"] == "validation"
].copy()

final_test_df = final_feature_df.loc[
    final_feature_df["SPLIT"] == "test"
].copy()

print("Rows before history filtering:", len(feature_df))
print(
    "Rows removed for incomplete history:",
    (~complete_history_mask).sum()
)
print("Final complete-history rows:", len(final_feature_df))

print("\nMissing values in history features:")
print(
    final_feature_df[history_feature_columns]
    .isna()
    .sum()
    .sum()
)

final_split_audit = (
    final_feature_df
    .groupby("SPLIT")
    .agg(
        rows=("NEXT_HOUR_CALLS", "size"),
        prediction_start=("NEXT_TIMESTAMP", "min"),
        prediction_end=("NEXT_TIMESTAMP", "max")
    )
    .reset_index()
)

final_split_audit["percentage"] = (
    final_split_audit["rows"]
    / len(final_feature_df)
    * 100
)

print("\nFinal split audit:")
display(final_split_audit)

print("\nChronological checks:")
print(
    "Train before validation:",
    final_train_df["NEXT_TIMESTAMP"].max()
    < final_validation_df["NEXT_TIMESTAMP"].min()
)

print(
    "Validation before test:",
    final_validation_df["NEXT_TIMESTAMP"].max()
    < final_test_df["NEXT_TIMESTAMP"].min()
)

print(
    "Unassigned rows:",
    (final_feature_df["SPLIT"] == "unassigned").sum()
)
```

```python
# Combine all leakage-safe candidate features
candidate_feature_columns = (
    calendar_columns
    + lag_columns
    + rolling_columns
    + priority_columns
    + call_type_columns
)

# Remove any accidental duplicate column names while preserving order
candidate_feature_columns = list(
    dict.fromkeys(candidate_feature_columns)
)

model_ready_df = final_feature_df[
    [
        "HOUR",
        "NEXT_TIMESTAMP",
        "SPLIT"
    ]
    + candidate_feature_columns
    + ["NEXT_HOUR_CALLS"]
].copy()

# Convert the feature matrix to numeric values for validation
numeric_feature_matrix = (
    model_ready_df[candidate_feature_columns]
    .apply(pd.to_numeric, errors="coerce")
)

future_or_target_columns = [
    column
    for column in candidate_feature_columns
    if column in {
        "NEXT_HOUR_CALLS",
        "NEXT_TIMESTAMP",
        "HOUR",
        "SPLIT"
    }
    or column.startswith("NEXT_")
]

constant_features = [
    column
    for column in candidate_feature_columns
    if numeric_feature_matrix[column].nunique(dropna=False) <= 1
]

print("Model-ready rows:", len(model_ready_df))
print("Candidate feature count:", len(candidate_feature_columns))

print(
    "Duplicate feature names:",
    len(candidate_feature_columns)
    - len(set(candidate_feature_columns))
)

print(
    "Non-numeric values after conversion:",
    numeric_feature_matrix.isna().sum().sum()
)

print(
    "Infinite feature values:",
    np.isinf(
        numeric_feature_matrix.to_numpy(dtype=float)
    ).sum()
)

print(
    "Target or future columns incorrectly included:",
    future_or_target_columns
)

print("Constant features:", constant_features)

print("\nRows by split:")
print(model_ready_df["SPLIT"].value_counts())

print("\nTarget summary by split:")
display(
    model_ready_df
    .groupby("SPLIT")["NEXT_HOUR_CALLS"]
    .describe()
)

print("\nFirst five model-ready rows:")
display(model_ready_df.head())
```
